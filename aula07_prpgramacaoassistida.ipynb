```python
numeros = [1,2,3] # Define uma lista de números inteiros.
soma = 0 # Inicializa a variável 'soma' com zero.

# Itera sobre cada número na lista 'numeros'.
for num in numeros:
  soma += num # Adiciona o número atual à variável 'soma'.

print(soma) # Imprime o valor final da soma.
```

Eh primo

```python
import math

def eh_primo(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0: # Todos os outros números pares não são primos
        return False
    # Verifica apenas divisores ímpares de 3 até a raiz quadrada de n
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

print(eh_primo(7)) # Deve ser True
print(eh_primo(2)) # Deve ser True
print(eh_primo(1)) # Deve ser False
print(eh_primo(4)) # Deve ser False
print(eh_primo(9)) # Deve ser False
print(eh_primo(17)) # Deve ser True
```
