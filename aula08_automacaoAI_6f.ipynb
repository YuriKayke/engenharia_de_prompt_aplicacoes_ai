```python
# Importa o módulo random para simular a temperatura do sistema
import random

def get_system_temperature():
    """
    Simula a obtenção da temperatura do sistema.
    Em um cenário real, esta função seria responsável por ler a temperatura
    de um sensor ou API de hardware.
    """
    # Gera um número aleatório entre 25.0 e 35.0 para simular a temperatura
    return random.uniform(25.0, 35.0)

def check_temperature_and_alert():
    """
    Verifica a temperatura do sistema e exibe um alerta de superaquecimento
    se a temperatura for maior que 30 graus Celsius.
    """
    # Obtém a temperatura simulada do sistema
    temperature = get_system_temperature()

    # Exibe a temperatura atual usando concatenação de strings
    print("Temperatura atual do sistema: " + str(round(temperature, 2)) + "°C")

    # Verifica se a temperatura ultrapassa o limite de 30 graus
    if temperature > 30:
        # Exibe um alerta de superaquecimento
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        # Informa que a temperatura está normal
        print("Temperatura do sistema está normal.")

# Executa a função para verificar a temperatura e alertar
check_temperature_and_alert()
```

# Programa de Organização de Arquivos com Criação de Amostras

Criar um programas que organize arquivos bagunçados em pastas automaticamente, baseando-se apenas na extensão do arquivo(ex. .pdf, .jpg, .xls).

Use a biblioteca os e shiutil.

Comente a execução durante o código.

Use 10 arquivos de cada tipo.

```python
import os
import shutil
import random

def create_sample_files(target_directory, num_files_per_type=10):
    """
    Cria arquivos de exemplo com diferentes extensões para demonstração.

    Args:
        target_directory (str): O caminho para o diretório onde os arquivos de exemplo serão criados.
        num_files_per_type (int): O número de arquivos a serem criados para cada tipo de extensão.
    """
    # Inicia a criação dos arquivos de exemplo e exibe uma mensagem informativa.
    print(f"\n--- Criando arquivos de exemplo no diretório: {target_directory} ---")
    # Cria o diretório de destino se ele não existir. 'exist_ok=True' evita erro se já existir.
    os.makedirs(target_directory, exist_ok=True)

    # Define as extensões de arquivo que serão usadas para os exemplos.
    extensions = ['pdf', 'jpg', 'png', 'txt', 'docx', 'xlsx', 'mp3', 'zip']

    # Itera sobre cada tipo de extensão definida.
    for ext in extensions:
        # Para cada extensão, cria o número especificado de arquivos.
        for i in range(num_files_per_type):
            # Gera um nome de arquivo único para evitar que arquivos com o mesmo nome se sobrescrevam.
            # Inclui a extensão, um índice e um número aleatório.
            filename = f"arquivo_{ext}_{i + 1}_{random.randint(1000, 9999)}.{ext}"
            # Constrói o caminho completo para o novo arquivo dentro do diretório de destino.
            filepath = os.path.join(target_directory, filename)
            # Abre o arquivo em modo de escrita ('w') e adiciona um conteúdo simples.
            with open(filepath, 'w') as f:
                f.write(f"Conteúdo de teste para {filename}")
            # Confirma a criação do arquivo.
            print(f"  Criado: {filename}")

    # Além dos arquivos com extensão, cria alguns arquivos sem extensão para teste.
    for i in range(num_files_per_type // 2):
        # Gera um nome de arquivo sem extensão, também com um número aleatório.
        filename = f"sem_extensao_{i + 1}_{random.randint(100, 999)}"
        # Constrói o caminho completo para o arquivo.
        filepath = os.path.join(target_directory, filename)
        # Abre e escreve um conteúdo no arquivo sem extensão.
        with open(filepath, 'w') as f:
            f.write(f"Conteúdo de teste para {filename}")
        # Confirma a criação do arquivo sem extensão.
        print(f"  Criado: {filename} (sem extensão)")

def organize_files_by_extension(directory_path):
    """
    Organiza arquivos em um diretório especificado, movendo-os para subpastas
    nomeadas de acordo com suas extensões.

    Args:
        directory_path (str): O caminho para o diretório a ser organizado.
    """
    # Inicia o processo de organização e exibe uma mensagem informativa.
    print(f"\n--- Iniciando a organização de arquivos em: {directory_path} ---")

    # Verifica se o diretório fornecido realmente existe. Se não, exibe um erro e encerra.
    if not os.path.isdir(directory_path):
        print(f"Erro: O diretório '{directory_path}' não foi encontrado. Por favor, verifique o caminho.")
        return

    # Lista todos os arquivos e subdiretórios presentes no diretório especificado.
    items = os.listdir(directory_path)
    print(f"Encontrados {len(items)} itens no diretório.")

    # Itera sobre cada item (arquivo ou diretório) encontrado.
    for item_name in items:
        # Constrói o caminho completo para o item atual.
        source_path = os.path.join(directory_path, item_name)

        # Verifica se o item é um diretório. Se for, ele é ignorado para não mover pastas já existentes.
        if os.path.isdir(source_path):
            print(f"  Ignorando diretório: {item_name}")
            continue

        # Divide o nome do arquivo em duas partes: o nome base e a extensão.
        # Ex: 'documento.pdf' se torna ('documento', '.pdf').
        # Ex: 'sem_extensao' se torna ('sem_extensao', '').
        name, extension = os.path.splitext(item_name)

        # Limpa e padroniza a extensão para ser usada como nome da pasta.
        # Remove o ponto inicial (ex: '.pdf' vira 'pdf') e converte para minúsculas.
        # Se não houver extensão (string vazia), define como 'no_extension'.
        clean_extension = extension.lower().replace('.', '')
        if not clean_extension:
            clean_extension = 'no_extension'

        # Define o nome da pasta de destino com base na extensão limpa.
        # Ex: 'pdf_files', 'jpg_files', 'no_extension_files'.
        destination_folder_name = f"{clean_extension}_files"
        # Constrói o caminho completo para a pasta de destino.
        destination_folder_path = os.path.join(directory_path, destination_folder_name)

        # Cria a pasta de destino se ela ainda não existir. 'exist_ok=True' evita um erro.
        os.makedirs(destination_folder_path, exist_ok=True)
        print(f"  Pasta de destino para '{clean_extension}' criada/existente: {destination_folder_name}")

        # Constrói o caminho completo para onde o arquivo será movido (dentro da pasta de destino).
        final_destination_path = os.path.join(destination_folder_path, item_name)

        try:
            # Move o arquivo do caminho original para o caminho de destino.
            shutil.move(source_path, final_destination_path)
            print(f"  Movido: '{item_name}' para '{destination_folder_name}'")
        except shutil.Error as e:
            # Captura erros específicos da biblioteca shutil, como se o arquivo já existir no destino.
            print(f"  AVISO: Não foi possível mover '{item_name}'. Motivo: {e}. O arquivo pode já existir no destino.")
        except Exception as e:
            # Captura quaisquer outros erros inesperados que possam ocorrer durante a movimentação.
            print(f"  ERRO: Ocorreu um problema ao mover '{item_name}': {e}")

    # Mensagem final indicando que a organização foi concluída.
    print("\n--- Organização de arquivos concluída! ---")

# --- Exemplo de Uso ---
# Esta seção demonstra como usar as funções definidas acima.
# ATENÇÃO: Altere 'meus_arquivos_teste' para o nome do diretório que você deseja usar.
# Certifique-se de que é um diretório de teste, pois os arquivos serão movidos!

# 1. Defina o diretório de teste onde os arquivos de exemplo serão criados e organizados.
test_dir = "./meus_arquivos_teste"

# 2. Chama a função para criar os arquivos de exemplo. Esta etapa é opcional se você já tem arquivos.
# Se você já tem arquivos, pode comentar ou remover esta linha.
create_sample_files(test_dir, num_files_per_type=10)

# 3. Executa a função principal de organização, passando o diretório de teste.
organize_files_by_extension(test_dir)
```

Desenvolver um script que consulte uma API publica (correios) e retorne as informações de endereço formatada em JSON. Use tratamento de exceção

### Consulta de Endereço via API Pública (ViaCEP)

Como a API direta dos Correios (SIGEP) não é publicamente acessível para consulta de CEPs por terceiros, utilizaremos a **API ViaCEP**. Esta é uma alternativa popular e aberta que consome e fornece dados dos Correios de forma prática. O script abaixo realizará a consulta, formatará a resposta em JSON e incluirá tratamento de exceções para diversos cenários de erro.

```python
import requests
import json

def consultar_cep_viacep(cep):
    """
    Consulta a API pública ViaCEP para obter informações de endereço a partir de um CEP.

    Args:
        cep (str): O Código de Endereçamento Postal (CEP) a ser consultado.

    Returns:
        dict or None: Um dicionário contendo as informações de endereço formatadas em JSON,
                      ou None se ocorrer um erro ou o CEP não for encontrado.
    """
    # 1. Limpa o CEP: Remove caracteres não numéricos para garantir o formato correto.
    cep_limpo = ''.join(filter(str.isdigit, cep))

    # 2. Valida o CEP: Verifica se o CEP tem exatamente 8 dígitos numéricos.
    if len(cep_limpo) != 8:
        print(f"Erro: O CEP '{cep}' é inválido. Um CEP deve conter 8 dígitos numéricos.")
        return None

    # 3. Constrói a URL da API ViaCEP.
    url = f"https://viacep.com.br/ws/{cep_limpo}/json/"

    try:
        # 4. Realiza a requisição GET para a API.
        print(f"Tentando consultar CEP: {cep_limpo} em {url}")
        response = requests.get(url, timeout=5) # Adiciona um timeout para evitar que a requisição trave indefinidamente

        # 5. Verifica o status da resposta HTTP.
        response.raise_for_status()  # Levanta um HTTPError para respostas de erro (4xx ou 5xx).

        # 6. Decodifica a resposta JSON.
        data = response.json()

        # 7. Verifica se a API retornou um indicativo de erro ou CEP não encontrado.
        if 'erro' in data and data['erro'] is True:
            print(f"CEP '{cep_limpo}' não encontrado na base de dados do ViaCEP.")
            return None

        # 8. Formata os dados de endereço em um dicionário para retorno.
        #    Usa .get() para evitar KeyError se algum campo estiver faltando na resposta.
        endereco_formatado = {
            "cep": data.get('cep'),
            "logradouro": data.get('logradouro'),
            "complemento": data.get('complemento'),
            "bairro": data.get('bairro'),
            "localidade": data.get('localidade'),
            "uf": data.get('uf'),
            "ibge": data.get('ibge'),
            "gia": data.get('gia'),
            "ddd": data.get('ddd'),
            "siafi": data.get('siafi')
        }

        return endereco_formatado

    # 9. Tratamento de Exceções: Captura e informa sobre diferentes tipos de erros.
    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP: {http_err} (Código de Status: {response.status_code}).")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de Conexão: Não foi possível conectar-se ao servidor ViaCEP. Verifique sua conexão com a internet. Detalhes: {conn_err}")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Erro de Timeout: A requisição excedeu o tempo limite de 5 segundos. Tente novamente mais tarde. Detalhes: {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"Erro na Requisição: Ocorreu um erro inesperado ao fazer a requisição. Detalhes: {req_err}")
    except json.JSONDecodeError as json_err:
        print(f"Erro de Decodificação JSON: A resposta da API não pôde ser interpretada como JSON válido. Detalhes: {json_err}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

    return None

# --- Exemplo de Uso do Script --- #
if __name__ == "__main__":
    print("\n--- Testando a Função de Consulta de CEP ---")

    # Solicita o CEP ao usuário
    cep_usuario = input("Por favor, digite o CEP que deseja consultar (ex: 01001-000): ")

    # Realiza a consulta com o CEP fornecido pelo usuário
    endereco_consultado = consultar_cep_viacep(cep_usuario)

    if endereco_consultado:
        print("\nEndereço encontrado (formato JSON):\n" + json.dumps(endereco_consultado, indent=4, ensure_ascii=False))
    else:
        print(f"\nFalha ao encontrar endereço ou CEP inválido para '{cep_usuario}'.")

    print("\n" + "="*60 + "\n")

    # Exemplos de testes adicionais podem ser incluídos aqui, se desejado,
    # mas o foco principal agora é a entrada do usuário.
```
